```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

# Two lists of sentences
sentences1 = [
    "The cat sits outside",
    "A man is playing guitar",
    "The new movie is awesome",
]

sentences2 = [
    "The dog plays in the garden",
    "A woman watches TV",
    "The new movie is so great",
]

# Compute embedding for both lists
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True)

# Compute cosine-similarities
cosine_scores = util.cos_sim(embeddings1, embeddings2)

# Output the pairs with their score
for i in range(len(sentences1)):
    print("{} \t\t {} \t\t Score: {:.4f}".format(
        sentences1[i], sentences2[i], cosine_scores[i][i]
    ))
```

```python
# code to create json from column PAparagraphs and save into column PApragraphs_Json

import pandas as pd
import json
import re
from openpyxl import load_workbook


df = pd.read_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx')


def extract_and_convert(row):
    #data = re.findall(r'\[([^]]+)\]\n([^[]+)', row)
    data = re.findall(r'\[([^]]+)\](?:\n)?\s*([^[]+)', row)
    return {
        "sentences": [{"id": d[0], "content": d[1].strip()} for d in data]
    }


df['PApragraphs_Json'] = df['PAparagraphs'].apply(extract_and_convert)


df['PApragraphs_Json'] = df['PApragraphs_Json'].apply(lambda x: json.dumps(x, ensure_ascii=False))


#with pd.ExcelWriter('gdrive/MyDrive/ClaimCiteRetrieval_ESR_based_new.xlsx', engine='openpyxl') as writer:
#    df.to_excel(writer, sheet_name='Sheet1', index=False)
df.to_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx', index=False)
```

```python
#Task 1,case 1.1 (step 1 to json)

from sentence_transformers import SentenceTransformer, util
import pandas as pd
import json

model = SentenceTransformer("all-MiniLM-L6-v2")


df = pd.read_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx')


for index, row in df.iterrows():
    col1_text = row['APclaim']


    col2_text = row['PAclaim']


    sentences1 = [col1_text]
    sentences2 = [col2_text]

    output_json = []



    embeddings1 = model.encode(sentences1, convert_to_tensor=True)
    embeddings2 = model.encode(sentences2, convert_to_tensor=True)
    cosine_scores = util.cos_sim(embeddings1, embeddings2).item()  # Convert the tensor to a float

    output_json.append({
        "sentences1": sentences1,
        "sentences2": sentences2,
        "score": "{:.4f}".format(cosine_scores)
    })


    output_json_str = json.dumps({"sentences": output_json}, indent=2)


    df.at[index, 'APclaimPAclaim-sim-minilml6v2-json'] = output_json_str


df.to_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx', index=False)
```

```python
#code for sentence embedding (Task 2) saving new json with score in a new column PAparagraphs_sim-mpnet

from sentence_transformers import SentenceTransformer, util
import pandas as pd
import json

model = SentenceTransformer("all-MiniLM-L6-v2")


df = pd.read_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx')


for index, row in df.iterrows():
    col1_text = row['APclaim']


    col2_json = json.loads(row['PApragraphs_Json'])


    sentences1 = [col1_text]

    output_json = []
    for item in col2_json['sentences']:
        sentences2_id = item['id']
        sentences2_content = item['content']

        embeddings1 = model.encode(sentences1, convert_to_tensor=True)
        embeddings2 = model.encode(sentences2_content, convert_to_tensor=True)
        cosine_scores = util.cos_sim(embeddings1, embeddings2).item()  # Convert the tensor to a float

        output_json.append({
            "id": sentences2_id,
            "content": sentences2_content,
            "score": "{:.4f}".format(cosine_scores)
        })


    output_json_str = json.dumps({"sentences": output_json}, indent=2)


    df.at[index, 'PAparagraphs_sim-minilml6v2'] = output_json_str


df.to_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx', index=False)
```

```python
#code to calculate top 3 scores (task 3) and save in new column PAparagraphs-sim-mpnet-top123
import pandas as pd
import json


df = pd.read_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx')


def get_top_3(row):
    try:
        output_json = json.loads(row['PAparagraphs_sim-minilml6v2'])
        sorted_scores = sorted(output_json['sentences'], key=lambda x: float(x['score']), reverse=True)
        top_3 = [[item['id'], "{:.4f}".format(float(item['score']))] for item in sorted_scores[:3]]
        return top_3
    except (json.JSONDecodeError, KeyError):
        return []


df['PAparagraphs-sim-minilml6v2-top123'] = df.apply(get_top_3, axis=1)
df.to_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx', index=False)
```

```python
#Task 4, 4.1(1)
import pandas as pd
import ast

df = pd.read_excel('gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx')


df['PAparagraphs-sim-minilml6v2-top123']  = df['PAparagraphs-sim-minilml6v2-top123'].apply(ast.literal_eval)

first_values = [item[0] for sublist in df['PAparagraphs-sim-minilml6v2-top123'] for item in sublist]

grouped_values = [f"[{value}]" for value in first_values]

other_column_values = df['PAcited'].tolist()

results_list = []
count_true_1 = count_true_2 = count_true_3 = 0


for i in range(0, len(grouped_values), 3):
    row_values = grouped_values[i:i+3]
    other_values = other_column_values[i//3]

    results = [str(value in other_values) for value in row_values]
    count_true_1 += int(results[0] == 'True')
    count_true_2 += int(results[1] == 'True')
    count_true_3 += int(results[2] == 'True')

    results_list.append(results)


    print("Values from 'top123':", row_values)
    print("Values from 'other_column':", other_values)
    print("Result:", ",".join(results))
    print()

total_rows = len(results_list)
av1 = (count_true_1 / total_rows) * 100
av2 = (count_true_2 / total_rows) * 100
av3 = (count_true_3 / total_rows) * 100

average_of_averages = (av1 + av2 + av3) / 3


print("Top1_minilml6v2_accuracy:", av1)
print("Top2_minilml6v2_accuracy:", av2)
print("Top3_minilml6v2_accuracy:", av3)

print("Average of av1, av2, av3:", average_of_averages)
```

```python
#Task 6 Part 1 , PACited which are present in Json
import pandas as pd
import json


file_path = 'gdrive/MyDrive/New-ClaimCiteRetrieval_ESR_based_24-02-2024.xlsx'
df = pd.read_excel(file_path)

# Initialize variables to keep track of minimum, maximum, and total average scores
min_average_score = float('inf')
max_average_score = float('-inf')
total_average_score = 0.0
num_rows = 0

for index, row in df.iterrows():
    values = row['PAcited']
    values_list = [value.strip('[ ]') for value in values.split(',')]

    json_data = row['PAparagraphs_sim-minilml6v2']
    json_obj = json.loads(json_data)

    scores = [float(sentence['score']) for sentence in json_obj['sentences'] if sentence['id'] in values_list]

    average_score = sum(scores) / len(scores) if scores else 0.0

    # Update minimum and maximum average scores
    min_average_score = min(min_average_score, average_score)
    max_average_score = max(max_average_score, average_score)

    # Update total average score and increment the row counter
    total_average_score += average_score
    num_rows += 1

    print(f"Values: {','.join(values_list)}, Average Score: {average_score}")

# Calculate overall average score
overall_average_score = total_average_score / num_rows if num_rows > 0 else 0.0

# Print minimum, maximum, and overall average scores
print()
print(f"minilm6v2_min_simi_cited_paragraph: {min_average_score}")
print(f"minilm6v2_max_simi_cited_paragraph: {max_average_score}")
print(f"minilm6v2_avg_simi_cited_paragraph: {overall_average_score}")
```
